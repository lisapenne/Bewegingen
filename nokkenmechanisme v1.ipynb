# Ontwerp van de nok
-> dit volgt uit de uitleg die Herman in zijn notebook heeft geplaatst

- Hefwet $S=f(\theta)$ met $S$ de gewenste lineaire verplaatsing van de volger, en $\theta$ de nok-hoek (in ons geval zijn dat de eerste 3 bulletpoints)

- Wij moeten het **vormprofiel van de nok** vinden => $R(\theta)$ van de steekkromme vinden: 

- **Aanpak:**
    - steekkromme _(pad van middelpunt volger tov nok)_ uitdrukken in poolcoördinaten

    - grootte drukhoek $\alpha$ checken

    - **ondersnijding** mag niet gebeuren _(Ondersnijding
onstaat als de nokrol te groot is, waardoor deze niet meer precies de contouren van de nok
kan volgen. De rol springt dan van het ene
contactpunt over naar het andere, waardoor
de heffunctie niet meer gevolgd word)_ 

    - nokprofiel afleiden: $x(\theta), y(\theta)$


INFO uit notebook:

- Type heffingswet: "stijgen-stilstand-dalen-stilstand" (we mogen onze opgave lichtjes aanpassen aan een specifieke applicatie die we als "trade-off" context wensen te gebruiken)

- hefwet die ontwikkeld moet worden is in het formaat van $f(\theta)$

- (ook uit PnO): probeer de drukhoek zo klein mogelijk te houden:

    - hoe "groter", hoe groter de kracht is waarmee je de nokvolger tegen de nok moet "duwen"
    
    Dit veroorzaakt onnodige slijtage en groter risico op vastlopen van het mechanisme (en nog wat zaken zoals een verlies in efficientie van overbrenging van het vermogen)

- kijk uit voor **spitse punten** (dit is nl een singulariteit (waar het systeem kapot gaat)); nadelen zijn: massive stress, rapid wear, and failure of the cam-follower system

    Dit kan vermeden worden indien: $$ \forall \theta: \quad C + f(\theta) + f''(\theta) \gt 0,$$

- we zullen meerdere keren moeten itereren om tot de beste oplossing te komen
